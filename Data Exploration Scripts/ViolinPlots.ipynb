```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
```

### Plotting allelic vector probability distributions

```python
in_dir = "../1000Genomes_Beta_Genotypes/"
```

```python
STR = "CSF1PO"
fname = in_dir + STR + "_Mem_Genotypes_Matrix.txt"
CSF1PO = pd.read_csv(fname, sep="\t")
CSF1PO = CSF1PO.rename(columns={CSF1PO.columns[0]: 'ID'})
col_names = CSF1PO.columns
#col_names = col_names[1:-1]
col_names
```

```python
# Melt DataFrame
melted_df = pd.melt(CSF1PO, 
                    id_vars=["ID", "Mem"], # Variables to keep
                    var_name="Alleles") # Name of melted variable

#melted_df
fig, ax = plt.subplots(figsize=(12,8))
sn.swarmplot(x='Alleles', y='value', data=melted_df, hue='Alleles', size=3)
#sn.catplot(data=melted_df, x='Alleles', y='value', kind='swarm', aspect=2.5)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.xlabel("Allele", fontsize=16)
plt.ylabel("Posterior Probability", fontsize=16)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), title="Allele")
plt.title("CSF1PO Allele Probability Vector Distributions\n1000 Genomes Project Data", size=18)
plt.show()
#plt.savefig("../AllelicDistributions/CSF1PO.png", dpi=500)
```
